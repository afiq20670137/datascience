# Jumlah Mobil berdasakan Model

```python
model = df.groupby('model').count()[['tax']].sort_values(by='tax',ascending=True).reset_index()
model = model.rename(columns={'tax':'numberOfCars'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=model['model'],y=model['numberOfCars'],color='royalblue')
plt.xticks(rotation=60)
```

# Ukuran mesin

```python
eng = df.groupby('engineSize').count()[['tax']].sort_values(by='tax').reset_index()
eng = eng.rename(columns={'tax':'count'})
```

```python

plt.figure(figsize=(15,5))
sns.barplot(x=eng['engineSize'],y=eng['count'],color='royalblue')
```

# seleksi fitur

```python
features = ['model,''year','mileage','tax','mpg','engineSize']
x=df[features]
y=df['price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x, y, random_state=70)
y_test.shape
```

# regresi linear

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred= lr.predict(x_test)
```

```python
score = lr.score(x_test, y_test)
print('Akurasi model regresi linear = ',score)
```

# membuat inputan model regresi linear

```python
input_data = np.array([[2019,500,145,30.2,2]])

prediction = lr.predict(input_data)
print('estimasi', prediction)
```

# membuat saving model

```python
import pickle

filename = 'estimasi_mobilmercy.sav'

pickle.dump(lr,open(filename,'wb'))
```
